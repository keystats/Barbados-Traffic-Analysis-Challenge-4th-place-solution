## Load Train Data

```python
import pandas as pd
df_train = pd.read_csv('cleaned_train.csv')
df_train.head()
```

```python
common_cols = ['responseId', 'view_label', 'videos', 'video_time', 'datetimestamp_start', 'datetimestamp_end', 'date', 'signaling', 'time_segment_id', 'cycle_phase']
df_enter = df_train[common_cols + ['ID_enter', 'congestion_enter_rating']].copy()
df_enter.rename(columns={'ID_enter': 'ID', 'congestion_enter_rating': 'congestion'}, inplace=True)
df_enter['type'] = 'enter'
print("DataFrame for 'enter' data created.")
```

```python
df_exit = df_train[common_cols + ['ID_exit', 'congestion_exit_rating']].copy()
df_exit.rename(columns={'ID_exit': 'ID', 'congestion_exit_rating': 'congestion'}, inplace=True)
df_exit['type'] = 'exit'
print("DataFrame for 'exit' data created.")
```

```python
df_train_transformed = pd.concat([df_enter, df_exit], ignore_index=True)
print("Combined DataFrame created.")
```

```python
df_train_transformed.head()
```

## Load Test Data



```python
df_test = pd.read_csv('TestInputSegments.csv')
df_test.head()
```

```python
common_cols = ['responseId', 'view_label', 'videos', 'video_time', 'datetimestamp_start', 'datetimestamp_end', 'date', 'signaling', 'time_segment_id', 'cycle_phase']
df_enter_test = df_test[common_cols + ['ID_enter', 'congestion_enter_rating']].copy()
df_enter_test.rename(columns={'ID_enter': 'ID', 'congestion_enter_rating': 'congestion'}, inplace=True)
df_enter_test['type'] = 'enter'
print("DataFrame for 'enter' test data created.")
```

```python
df_exit_test = df_test[common_cols + ['ID_exit', 'congestion_exit_rating']].copy()
df_exit_test.rename(columns={'ID_exit': 'ID', 'congestion_exit_rating': 'congestion'}, inplace=True)
df_exit_test['type'] = 'exit'
print("DataFrame for 'exit' test data created.")
```

```python
df_test_transformed = pd.concat([df_enter_test, df_exit_test], ignore_index=True)
print("Combined test DataFrame created.")
```

```python
df_test_transformed.head()
```

```python
df_train_features = pd.read_csv('final_merged_train_features.csv')
df_train_features.head()
```

```python
df_train_features_processed = df_train_features.drop(columns=['ID', 'videos'])
df_train_features_processed.rename(columns={'Direction': 'type'}, inplace=True)
df_train_features_processed['type'] = df_train_features_processed['type'].str.lower()
df_train_features_processed['type'] = df_train_features_processed['type'].replace({'entry': 'enter'})
print("df_train_features_processed created and first few rows displayed.")
df_train_features_processed.head()
```

```python
df_train_merged = pd.merge(df_train_transformed, df_train_features_processed, on=['time_segment_id', 'view_label', 'type'], how='left')
print("Re-merged df_train_merged created and first few rows displayed.")
df_train_merged.head()
```

```python
df_test_features = pd.read_csv('final_merged_test_features.csv')
df_test_features.head()
```

```python
df_test_features_processed = df_test_features.drop(columns=['ID', 'videos'])
df_test_features_processed.rename(columns={'Direction': 'type'}, inplace=True)
df_test_features_processed['type'] = df_test_features_processed['type'].str.lower()
print("df_test_features_processed created and first few rows displayed.")
df_test_features_processed.head()
```

```python
df_test_features_processed['type'] = df_test_features_processed['type'].replace({'entry': 'enter'})
df_test_merged = pd.merge(df_test_transformed, df_test_features_processed, on=['time_segment_id', 'view_label', 'type'], how='left')
print("df_test_merged created and first few rows displayed.")
df_test_merged.head()
```

```python
df_train_merged.to_csv('df_train_merged.csv', index=False)
df_test_merged.to_csv('df_test_merged.csv', index=False)
```
