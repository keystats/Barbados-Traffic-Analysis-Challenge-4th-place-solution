# MERGING VIDEO AND BRIGHTNESS FEATURES FOR TRAIN

```python
import pandas as pd
import glob
import os

def merge_and_enrich_features(traffic_file_pattern, brightness_file, output_filename="final_merged_train_features.csv"):
    """
    Merges all extracted traffic feature CSVs and enriches them by joining
    with the contextual brightness feature CSV.

    Args:
        traffic_file_pattern (str): The file pattern to match all traffic feature files
                                    (e.g., 'extracted_traffic_features_*.csv').
        brightness_file (str): The filename for the brightness context data
                               (e.g., 'extracted_brightness_features.csv').
        output_filename (str): The name for the final combined CSV file.
    """

    # --- Step 1: Concatenate all Traffic Feature Files ---

    # Use glob to find all files matching the pattern
    traffic_files = glob.glob(traffic_file_pattern)

    if not traffic_files:
        print(f"CRITICAL ERROR: No traffic feature files found matching pattern '{traffic_file_pattern}'.")
        print("Please ensure your files are uploaded and the pattern is correct.")
        return

    print(f"Found {len(traffic_files)} traffic feature files to concatenate...")

    all_traffic_data = []

    # Read each file and store the DataFrame
    for f in traffic_files:
        try:
            df = pd.read_csv(f)
            all_traffic_data.append(df)
            print(f"  -> Successfully loaded {f}")
        except Exception as e:
            print(f"WARNING: Could not load file {f}. Skipping. Error: {e}")

    if not all_traffic_data:
        print("CRITICAL ERROR: No valid traffic data could be loaded. Exiting merge.")
        return

    # Concatenate all DataFrames into one massive traffic feature set
    traffic_df_raw = pd.concat(all_traffic_data, ignore_index=True)
    print(f"\n✅ Merged all traffic features into a single DataFrame with {len(traffic_df_raw)} rows.")


    # --- Step 2: Enrich with Brightness Context ---

    # Load the brightness data
    if not os.path.exists(brightness_file):
        print(f"\nCRITICAL ERROR: Brightness file '{brightness_file}' not found.")
        print("The final output will only contain traffic features.")
        final_df = traffic_df_raw
    else:
        brightness_df = pd.read_csv(brightness_file)
        print(f"Loaded brightness features with {len(brightness_df)} unique segments.")

        # We join on the common video identifiers: 'time_segment_id' and 'videos'.

        final_df = pd.merge(
            traffic_df_raw,
            brightness_df[['time_segment_id', 'videos', 'Avg_Brightness_Value', 'Day_Night_Label']],
            on=['time_segment_id', 'videos'],
            how='left' # Keep all traffic rows, but add brightness where available
        )


        print(f"✅ Successfully joined brightness context. Final DataFrame has {len(final_df)} rows.")


    # --- Step 3: Save the Final Merged File ---
    final_df.to_csv(output_filename, index=False)
    print("\n" + "="*80)
    print(f"🎉 FINAL MERGE COMPLETE!")
    print(f"Output saved to: **{output_filename}**")
    print(f"Total Rows: {len(final_df)}")
    print("="*80)


if __name__ == "__main__":
    # The pattern matches all files starting with the prefix from your notebooks
    TRAFFIC_PATTERN = 'extracted_traffic_features_A*.csv'
    BRIGHTNESS_FILE = 'extracted_brightness_features_train.csv'

    merge_and_enrich_features(
        traffic_file_pattern=TRAFFIC_PATTERN,
        brightness_file=BRIGHTNESS_FILE
    )
```
